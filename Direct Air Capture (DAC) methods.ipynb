```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Fetching Data (Structured Dataset)
data = {
    'Method': ['Solid Sorbent DAC', 'Liquid Solvent DAC', 'Cryogenic DAC', 'Electrochemical DAC'],
    'Energy Consumption (kWh/tonne CO₂)': [2000, 1500, 2500, 1750]
}

# Step 2: Processing the Data (Create DataFrame)
df = pd.DataFrame(data)

# Display the dataset
print("Fetched Data:")
display(df)

# Step 3: Analyzing the Data (Basic Statistics)
print("\nStatistical Summary:")
stats = df.describe()
display(stats)

# Step 4: Visualizing the Data (Bar Chart)
plt.figure(figsize=(10, 6))
plt.bar(df['Method'], df['Energy Consumption (kWh/tonne CO₂)'], color=['blue', 'green', 'red', 'orange'])
plt.xlabel('Direct Air Capture Method', fontsize=12, fontweight='bold')
plt.ylabel('Energy Consumption (kWh/tonne CO₂)', fontsize=12, fontweight='bold')
plt.title('Energy Consumption of Different Direct Air Capture (DAC) Methods', fontsize=14, fontweight='bold')
plt.xticks(rotation=15)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
